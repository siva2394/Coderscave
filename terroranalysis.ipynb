```python
df.rename(columns={
    'iyear': 'Year',
    'imonth': 'Month',
    'iday': 'Day',
    'country_txt': 'Country',
    'region_txt': 'Region',
    'city': 'City',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'attacktype1_txt': 'AttackType',
    'nkill': 'Fatalities',
    'nwound': 'Wounded'
}, inplace=True)
selected_columns = [
    'Year', 'Month', 'Day', 'Country', 'Region', 'City', 'Latitude', 'Longitude',
    'AttackType', 'Fatalities', 'Wounded'
]
df = df[selected_columns].copy()
print(df.head())
```

```python
df['Fatalities']=df['Fatalities'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
df['Casualities']=df['Fatalities']+df['Wounded']
df.isnull().sum()
```

What We Understood

The dataset spans from 1970 to 2017, covering various years, months, and days. There seems to be an even distribution across the years, with a range of incidents reported each month and day.

The latitude and longitude values fall within the expected ranges for global coordinates. However, there are extreme values, such as negative longitude and latitude below -90 and above 90, which might need verification or cleaning.

There's a wide range in the number of casualties reported per incident. The majority of incidents report no or very few casualties, but some incidents stand out with high casualties, potentially indicating severe attacks or events.

 Investigating incidents with extremely high casualties or unusual geographical coordinates could provide valuable insights into significant events or errors in the data.

```python
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
df.isnull().sum()
```

```python
# Assuming your DataFrame is named 'df'

# Grouping by 'Country' and 'Region', filling missing 'City' values with mode
df['City'] = df.groupby(['Country', 'Region'])['City'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else x.mode()))

```

```python

yearly_attacks = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(yearly_attacks.index, yearly_attacks.values, marker='o', linestyle='-')
plt.title('Number of Terrorist Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
attacks_per_region = df['Region'].value_counts()
plt.figure(figsize=(12, 6))
attacks_per_region.plot(kind='bar', color='skyblue')
plt.title('Number of Terrorist Attacks per Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

top_regions = df['Region'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 6))
sns.countplot(y='Region', data=df[df['Region'].isin(top_regions)],hue='Region', palette='rocket', order=top_regions)
plt.title('Top Regions Most Affected by Terrorism')
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.show()
```

```python
casualties_region = df.groupby('Region')[['Fatalities', 'Wounded']].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=casualties_region.melt(id_vars='Region', var_name='Casualties'), x='Region', y='value', hue='Casualties')
plt.title('Average Fatalities and Wounded per Region')
plt.xlabel('Region')
plt.ylabel('Average Count')
plt.xticks(rotation=45)
plt.legend(title='Casualties')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(x='Region', data=df,hue='Region', palette='viridis')
plt.title('Distribution of Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()

top_countries = df['Country'].value_counts().nlargest(15).index
plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df[df['Country'].isin(top_countries)],hue='Country', palette='viridis', order=top_countries)
plt.title('Top Countries Most Affected by Terrorism')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
attack_type_region = df.groupby(['Region', 'AttackType']).size().unstack()
plt.figure(figsize=(12, 18))
attack_type_region.plot(kind='bar', stacked=True)
plt.title('Distribution of Attack Types per Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Fatalities'], df['Wounded'], alpha=0.5, color='red')
plt.title('Fatalities vs Wounded')
plt.xlabel('Fatalities')
plt.ylabel('Wounded')
plt.tight_layout()
plt.show()
```
